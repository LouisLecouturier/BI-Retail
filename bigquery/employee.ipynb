```python
from google.cloud import bigquery
import pandas as pd
```

```python
# Initialiser le client BigQuery
client = bigquery.Client.from_service_account_json('key.json')
```

```python
def import_table(table : str):
    query = f"""
    SELECT *
    FROM `power-bi-m2.test_csv.{table}`
    """

    return client.query(query).to_dataframe()

employee = import_table('Employee')

employee = employee.drop(columns=['supervised_by'])

employee[['first name', 'last name']] = employee['name'].str.split(expand=True)

employee = employee.drop(columns=['name'])

display(employee)
```
