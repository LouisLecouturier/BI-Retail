```python
from google.cloud import bigquery
import pandas as pd

from utils.import_table import import_table
from utils.export import export_table
```

```python
# Initialiser le client BigQuery
client = bigquery.Client.from_service_account_json('../../key.json')
```

```python
sms = import_table(client, 'StockMovementStatus') # Remplir le nom de la table à importer
sm = import_table(client, 'StockMovement') # Remplir le nom de la table à importer
so = import_table(client, 'SalesOrder') # Remplir le nom de la table à importer
po = import_table(client, 'PurchaseOrder') # Remplir le nom de la table à importer
```

```python
display(sms) 
print(len(sms['date'].unique())) # Afficher les valeurs uniques de la colonne stock_movement_id
print(len(sm['date'].unique())) # Afficher les valeurs uniques de la colonne stock_movement_id
print(len(so['date'].unique())) # Afficher les valeurs uniques de la colonne stock_movement_id
print(len(po['date'].unique())) # Afficher les valeurs uniques de la colonne stock_movement_id

```

```python
# dates is a df of all the dates in sms, sm, so, po
dates = pd.concat([sms['date'], sm['date'], so['date'], po['date']]).drop_duplicates().sort_values().reset_index(drop=True)

# Afficher le nombre de dates uniques
print(len(dates))

# Ajouter une colonne id dans dates, en commençant à 1
dates = dates.reset_index(drop=True).reset_index()
dates['index'] = dates['index'] + 1
dates = dates.rename(columns={'index': 'id'})

# Afficher les dates uniques
display(dates)
```

```python
# Variables pour BigQuery
dataset_name = "warehouse_test" # Nom du dataset à créer/exporter
table_name = "date" # Nom de la table à créer

export_table(client, dates, dataset_name, table_name)
```
